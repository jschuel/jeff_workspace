```python
import analysis as a
import numpy as np
import pandas as pd
import root_pandas as rp
import matplotlib.pyplot as plt
import ROOT
import array
import seaborn as sns
```

```python
df = a.make_heuristic_dataframe(12, "LER", "iiwi")
```

```python
df = df.dropna()
```

```python
nb7890 = df.loc[(df['Nb']==789) & (df['knob']==0)]
nb7891 = df.loc[(df['Nb']==789) & (np.abs(df['knob'])==1)]
nb78919 = df.loc[(df['Nb']==789) & (np.abs(df['knob'])==1.9)]
nb15760 = df.loc[df['Nb']==1576 & (df['knob']==0)]
nb15761 = df.loc[(df['Nb']==1576) & (np.abs(df['knob'])==1)]
nb157619 = df.loc[(df['Nb']==1576) & (np.abs(df['knob'])==1.9)]
df
```

```python
plt.rc('legend', fontsize=16)
plt.rc('xtick', labelsize=24)
plt.rc('ytick', labelsize=24)
plt.rc('axes', labelsize=26)
plt.rc('axes', titlesize=26)
fig, ax = plt.subplots(1,1, figsize=(10,8))
x = array.array('d', df['x'])
y = array.array('d', df['y'])
x_err = array.array('d', df['x_err'])
y_err = array.array('d', df['y_err'])
gr = ROOT.TGraphErrors(len(x), x, y, x_err, y_err)
fit_pars = gr.Fit("pol1", "SB")
a = fit_pars.Parameter(0)
b = fit_pars.Parameter(1)
ax.errorbar(nb7890['x']/1000, nb7890['y']/1000, nb7890['y_err']/1000, nb7890['x_err']/1000, 'o', markeredgecolor = 'black', markersize = 12, lw = 2, color = 'blue', alpha = 0.9, label = r'ECK = $0$')
ax.errorbar(nb7891['x']/1000, nb7891['y']/1000, nb7891['y_err']/1000, nb7891['x_err']/1000, 's', markeredgecolor = 'black', markersize = 12, lw = 2, color = 'red', alpha = 0.9, label = r'ECK = $\pm 1$')
ax.errorbar(nb78919['x']/1000, nb78919['y']/1000, nb78919['y_err']/1000, nb78919['x_err']/1000, '^', markeredgecolor = 'black', markersize = 12, lw = 2, color = 'green', alpha = 0.9, label = r'ECK = $\pm 1.9$')
ax.errorbar(nb15760['x']/1000, nb15760['y']/1000, nb15760['y_err']/1000, nb15760['x_err']/1000, 'o', markeredgecolor = 'black', markersize = 12, fillstyle='none', color = 'blue', alpha = 0.9)
ax.errorbar(nb15761['x']/1000, nb15761['y']/1000, nb15761['y_err']/1000, nb15761['x_err']/1000, 's', markeredgecolor = 'black', markersize = 12, fillstyle='none', color = 'red', alpha = 0.9)
ax.errorbar(nb157619['x']/1000, nb157619['y']/1000, nb157619['y_err']/1000, nb157619['x_err']/1000, '^', markeredgecolor = 'black', markersize = 12, fillstyle='none', color = 'green', alpha = 0.9)
ax.text(0.95, 0.07, 'Beam Gas',
        verticalalignment='bottom', horizontalalignment='right', transform=ax.transAxes, fontsize=25)
ax.text(0.94, 0.24, 'Touschek',
        verticalalignment='bottom', horizontalalignment='right', transform=ax.transAxes, fontsize=25)
x_plt = np.linspace(0,40,41)
ax.plot(x_plt, a/1000+b*x_plt, '--', color = 'black', label = 'Fit')
ax.fill_between(x_plt,0,a/1000,color = 'cyan', alpha = 0.3)
ax.fill_between(x_plt,a/1000,a/1000+b*x_plt, color = 'magenta', alpha = 0.3)
ax.set_xlim(0,40)
ax.set_ylim(0,4)
ax.set_ylabel(r'$\frac{\mathcal{O}}{IPZ_e^2}$ [Hz mA$^{-1}$ mPa$^{-1}$]')
ax.set_xlabel(r'$\frac{I}{P\sigma_yn_bZ_e^2}$ [mA mPa$^{-1}$ $\mathrm{\mu}$m$^{-1}$]')
plt.grid()
plt.legend(loc = 'upper left')
plt.savefig('combined_heuristic.jpg', bbox_inches='tight',dpi=300)
```
