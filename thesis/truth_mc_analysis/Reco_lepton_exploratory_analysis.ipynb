```python
import numpy as np
import pandas as pd
import root_pandas as rp
import matplotlib.pyplot as plt
```

```python
df = rp.read_root("/home/jeff/data/thesis_data/reco_lepton_ntuples/reco_leptons.root")
```

```python
#df = df.loc[df['dr']<100]
#df = df.loc[df['E']<1000]
#df = df.loc[np.abs(df['z'])<5]
df = df.loc[(df['CMS_p']>1.4) & (df['CMS_p']<1.9)]
df = df.loc[df['CMS_ImpactXY']<.3]
df = df.loc[df['CMS_dr']<.1]
#df = df.loc[np.abs(df['CMS_dz'])<100]
```

```python
sig = df.loc[df['isSignal']==1]
bg = df.loc[df['isSignal']==0]
```

```python
plt.figure(figsize = (15,100))
i=1
for col in df.columns:
    plt.subplot(24,3,i)
    plt.hist(sig[col],histtype='step', label = 'signal',bins = 100)
    plt.hist(bg[col],histtype='step', label = 'bg', bins = 100)
    plt.title(col)
    plt.yscale("Log")
    i+=1
```
